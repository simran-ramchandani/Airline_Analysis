```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
data=pd.read_csv("mark.csv")
```

```python
data
```

```python
data.drop(columns="Unnamed: 0")
```

```python
data.drop_duplicates(keep="first")
```

```python
data["Customer Type"].unique()
```

```python
data["Type of Travel"].unique()
```

```python
data["Class"].unique()
```

```python
data["satisfaction"].unique()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
la=LabelEncoder()
```

```python
obj_data=data.select_dtypes(include="object")
```

```python
data["Customer Type"].unique()
```

```python
la.fit_transform(data['Customer Type'])
```

```python
for col in obj_data.columns:
    data[col]=la.fit_transform(data[col])
```

```python
x=data.iloc[:,:-1]
y=data["satisfaction"]
```

```python
data
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.2,random_state=62)
```

```python
x_train
```

```python
x_test
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='mean')  # Or 'median', 'most_frequent'
x_train = imputer.fit_transform(x_train)
x_test = imputer.fit_transform(x_test)
```

LinearRegression

```python
from sklearn.linear_model import LinearRegression
```

```python
li=LinearRegression()
```

```python
li.fit(x_train,y_train)
```

```python
pred_point=li.predict(x_test)
```

```python
pred_point = np.round(pred_point)
pred_point = pred_point.astype(int)
```

```python
y_test
```

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
```

```python
r2_score(pred_point,y_test)
```

```python
pred_point
```

```python
mean_absolute_error(pred_point,y_test)
```

```python
np.sqrt(mean_squared_error(pred_point,y_test))
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lo = LogisticRegression()
```

```python
lo.fit(x_train,y_train)
```

```python
lo_pred = lo.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score,confusion_matrix,classification_report
```

```python
confusion_matrix(lo_pred,y_test)
```

```python
print(classification_report(lo_pred,y_test))
```

```python
print('Logistic model stats are :')
print(f'accuracy_score : {accuracy_score(lo_pred,y_test)}'),
print(f'precision_score : {precision_score(lo_pred,y_test)}')
print(f'recall_score : {recall_score(lo_pred,y_test)}')
print(f'f1_score : {f1_score(lo_pred,y_test)}')
```

Naive byse theorem

```python
from sklearn.naive_bayes import GaussianNB,BernoulliNB,MultinomialNB
```

```python
gnb = GaussianNB()
bnb = BernoulliNB()
mnb = MultinomialNB()
```

```python
gnb.fit(x_train,y_train)
bnb.fit(x_train,y_train)
mnb.fit(x_train,y_train)
```

```python
y_pred =gnb.predict(x_test)
print(classification_report(y_test, y_pred))

accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy Score: {accuracy}")
```

```python
y_pred =bnb.predict(x_test)
print(classification_report(y_test, y_pred)) 

accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy Score: {accuracy}")
```

```python
y_pred = mnb.predict(x_test)

print(classification_report(y_test, y_pred))
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy Score: {accuracy}")
```

KNN K nearest neighbour
