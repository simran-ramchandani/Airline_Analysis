```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
data=pd.read_csv("mark.csv")
```

```python
data.drop(columns="Unnamed: 0",inplace=True)
```

```python
satisfaction_counts = data["satisfaction"].value_counts()

plt.pie(satisfaction_counts, labels=satisfaction_counts.index, autopct='%1.1f%%', startangle=90,colors=['#4682B4', 'orange'])
plt.axis('equal') 
plt.show()
```

```python
satisfaction_counts = data["Customer Type"].value_counts()
total_counts = satisfaction_counts.sum()
percentages = (satisfaction_counts / total_counts) * 100

x = percentages.index
y = percentages.values

plt.xlabel('Customer Type')
plt.ylabel('Percentage')
plt.title('Satisfaction Distribution (Percentage)')
bars = plt.bar(x, y, color=['#4682B4', 'orange'])

for bar, percentage in zip(bars, y):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height(), f'{percentage:.1f}%', 
             ha='center', va='bottom', fontsize=10)

plt.xticks(rotation=45)
plt.show()
```
